```python
# CANBK ARIMA Assignment Visualization Script

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA

# Style
sns.set(style="darkgrid")

# Generate dummy stock data (1 year)
np.random.seed(42)
dates = pd.date_range(start="2024-01-01", periods=365, freq='D')
price = 300 + np.cumsum(np.random.normal(0, 2, 365))  # random walk

df = pd.DataFrame({"Date": dates, "Close": price})
df.set_index("Date", inplace=True)

# -------- 1. Closing Price Trend --------
plt.figure(figsize=(10,5))
plt.plot(df['Close'], color='blue')
plt.title("CANBK Closing Price Trend (1 Year)")
plt.xlabel("Date")
plt.ylabel("Price")
plt.savefig("1_price_trend.png")
plt.show()

# -------- 2. Rolling Mean & Std --------
rolling_mean = df['Close'].rolling(window=30).mean()
rolling_std = df['Close'].rolling(window=30).std()

plt.figure(figsize=(10,5))
plt.plot(df['Close'], label='Original')
plt.plot(rolling_mean, label='Rolling Mean', color='red')
plt.plot(rolling_std, label='Rolling Std', color='green')
plt.legend()
plt.title("Rolling Mean & Standard Deviation")
plt.savefig("2_rolling_stats.png")
plt.show()

# -------- 3. ADF Test --------
result = adfuller(df['Close'])
print("ADF Statistic:", result[0])
print("p-value:", result[1])

# -------- 4. Differenced Data --------
df_diff = df['Close'].diff().dropna()

plt.figure(figsize=(10,5))
plt.plot(df_diff, color='purple')
plt.title("Differenced Series (Stationary)")
plt.savefig("3_differenced.png")
plt.show()

# -------- 5. ACF & PACF --------
fig, ax = plt.subplots(1,2, figsize=(12,4))
plot_acf(df_diff, ax=ax[0])
plot_pacf(df_diff, ax=ax[1])
plt.savefig("4_acf_pacf.png")
plt.show()

# -------- 6. ARIMA Model --------
model = ARIMA(df['Close'], order=(1,1,1))
model_fit = model.fit()

# -------- 7. Forecast --------
forecast = model_fit.forecast(steps=30)

plt.figure(figsize=(10,5))
plt.plot(df['Close'], label='Historical')
plt.plot(pd.date_range(df.index[-1], periods=30, freq='D'),
         forecast, label='Forecast', color='red')
plt.legend()
plt.title("30-Day Forecast (ARIMA)")
plt.savefig("5_forecast.png")
plt.show()
```
